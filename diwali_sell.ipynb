```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt # visualizing data
import seaborn as sns
```

```python
df = pd.read_csv('Diwali Sales Data.csv', encoding= 'unicode_escape')
print(df.to_string())
```

```python
df.shape
```

```python
df.head()
```

```python
df.info()
```

```python
df.drop(['Status', 'unnamed1'], axis=1, inplace=True) # drop column 'Status', 'unnamed1'
```

```python
df.info()
```

```python
pd.isnull(df).sum() # column wise null value calculator
```

```python
df.dropna(inplace=True) # drop null values from the columns
```

```python
pd.isnull(df).sum()
```

```python
df['Amount'] = df['Amount'].astype('int') # change the data type of Amount(float64) into int
```

```python
df['Amount'].dtypes # check the type of column
```

```python
df.columns # show all the columns
```

```python
# Renaming the column
df.rename(columns= {'Marital_Status':'Shaadi'},inplace=True) # inplace = True(change in the original file)
```

```python
df.columns
```

```python
# df.describe() # method returns description of the data in the DataFrame (i.e. count, mean, std, etc)
df[['Age', 'Orders', 'Amount']].describe() # selective describe
```

```python
# ax=sns.countplot(x="Gender", data=df)

# Custom colors for categories
custom_colors = {'M': 'skyblue', 'F': 'lightcoral'}

# Plot with custom palette
ax = sns.countplot(x='Gender', data=df, palette=custom_colors)
print(ax)
# print(ax.containers)
for bar in ax.containers:
    ax.bar_label(bar)

# plt.show()
```

```python
sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x = 'Gender',y= 'Amount' ,data = sales_gen)
```

```python
ax = sns.countplot(data = df, x = 'Age Group', hue = 'Gender')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# Total Amount vs Age Group
sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x = 'Age Group',y= 'Amount' ,data = sales_age)
```

```python
# Total number of orders from top 10 states

sales_age = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,5)})

sns.barplot(x = 'State',y= 'Orders' ,data = sales_age,)
```

```python
# Total amount/sales from top 10 states.

# Total number of orders from top 10 states

sales_age = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x = 'State',y= 'Amount' ,data = sales_age)
```

```python
# Marital state

ax = sns.countplot(data = df, x = 'Shaadi')

sns.set(rc={'figure.figsize':(5,3)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Shaadi', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

# sns.set(rc={'figure.figsize':(5,3)})
# sns.barplot(data = sales_state, x = 'Shaadi',y= 'Amount', hue='Gender')

# Replace 0 and 1 with readable labels
sales_state['Shaadi'] = sales_state['Shaadi'].map({0: 'Not Married', 1: 'Married'})

# Plot
sns.set(rc={'figure.figsize':(5,3)})
sns.barplot(data=sales_state, x='Shaadi', y='Amount', hue='Gender')

plt.title("Sales by Marital Status and Gender")
plt.tight_layout()
plt.show()
```

```python
# occupation 
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(data = df, x = 'Occupation')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Occupation',y= 'Amount')
```

```python
# product category
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(data = df, x = 'Product_Category',)

for bars in ax.containers:
    ax.bar_label(bars)
plt.xticks(rotation=45)
plt.tight_layout()
```

```python
sales_state = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Product_Category',y= 'Amount')
plt.show()
```

```python
sales_state = df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Product_ID',y= 'Orders')
plt.show()
```

```python
# top 10 sold products
fig1, ax1 = plt.subplots(figsize=(12,7))
df.groupby('Product_ID')['Orders'].sum().nlargest(10).sort_values(ascending=False).plot(kind='bar')
plt.show()
```
